```python
from pathlib import Path
import h5py
import numpy as np
import matplotlib.pyplot as plt

root = Path("../EELS_single/training_set")

data = {}
for fp in sorted(root.glob("CONFMAT_*.hdf5")):
    element = fp.stem.replace("CONFMAT_", "")
    with h5py.File(fp, "r") as f:
        data[element] = np.array(f["spectra"], dtype=np.float32)

print({k: v.shape for k, v in data.items()})
```

```python
from pathlib import Path
import h5py
import numpy as np
import matplotlib.pyplot as plt

for fp in [
    Path("../EELS/trainingset/TRAIN_S.hdf5"),
    #Path("../EELS/single_element_spectra/CONFMAT_Ag.hdf5"),
]:
    with h5py.File(fp, "r") as f:
        x = np.array(f["spectra"], dtype=np.float32)
    print(fp.name, x.shape, x.min(), x.max(), x.mean(), x.std())

    plt.figure(figsize=(12,4))
    plt.plot(x[60])
    plt.title(fp.name)
    plt.show()

    plt.figure(figsize=(12,4))
    plt.plot(x.mean(axis=0))
    plt.title(fp.name + " mean")
    plt.show()
```
